```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X = df1.drop('Chances Of Admit',axis=1)
y = df1['Chances Of Admit']
```

```python
x=df1.iloc[:,:-1].values
y=df1.iloc[:,-1].values
```

```python
x.shape
```

```python
y.shape
```

```python
y=y.reshape(1903,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import PowerTransformer
scaler =PowerTransformer()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
y_train=scaler.fit_transform(y_train)
y_test=scaler.transform(y_test)
```

```python
import tensorflow as tf
ann=tf.keras.models.Sequential()
```

```python
from sklearn.model_selection import RandomizedSearchCV

hyper_parameter_grid={
    'n_estimators':[100,200,300,400,500,600,700,800,900,1000],
    'max_depth':[2,4,6,8,10,12,14],
    'min_samples_split':[5,15,25,35,45,55],
    'max_features':['auto','sqrt','log2'],
    'min_samples_leaf':[2,4,6,8]
}

search = RandomizedSearchCV(estimator=ann, param_distributions=hyper_parameter_grid,
                              cv=5,n_iter=20,scoring='neg_mean_absolute_error',verbose=3,n_jobs=1,random_state=121)
```

```python
ann.add(tf.keras.layers.Dense(units=7,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
```

```python
ann.compile(optimizer='adam',loss='mean_squared_error')
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
y_pred=ann.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
acc=r2_score(y_test,y_pred)
acc
```

```python
y_pred=ann.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```
